```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel (r'SaleData.xlsx')
data.head()
```

### A. Find the total sale amount region wise, manager wise, sales man wise. 

```python
table = pd.pivot_table(data, values='Sale_amt',columns=['Region'], aggfunc=np.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(data, values='Sale_amt',columns=["Manager"], aggfunc=np.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(data, values='Sale_amt',columns=['Region',"Manager"], aggfunc=np.sum, fill_value=0)
table
```

### B. Find the item wise unit sold and the region wise total sale. 

```python
table = pd.pivot_table(data, values=["Units",'Sale_amt'], index=['Item', 'Region'],columns=[], aggfunc=np.sum, fill_value=0)
table
```
