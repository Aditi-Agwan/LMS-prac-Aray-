```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel (r'SaleData.xlsx')
data.head()
```

### A. Find the total sale amount region wise, manager wise, sales man wise. 

```python
table = pd.pivot_table(data,index="Region",values="Sale_amt", aggfunc = np.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(data, index="Manager",values="Sale_amt", aggfunc = np.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(data, index="SalesMan",values="Sale_amt", aggfunc = np.sum, fill_value=0)
table
```

### B. Find the item wise unit sold and the region wise total sale. 

```python
table = pd.pivot_table(data, values=["Units"], index=['Region','Item'],columns=[], aggfunc=[np.sum], fill_value=0,margins=True)
table
```

### C. Count the manager wise sale and mean value of sale amount and also display the sum of all sale amount at the bottom

```python
table = pd.pivot_table(data,index=["Manager"],values=["Sale_amt"],aggfunc=[np.mean,len],fill_value=0,margins=True)
table
```

### D. Find the total sale amount region wise, manager wise, sales man wise where Manager = "Timothy"

```python
table = pd.pivot_table(data,index=["Region","Manager","SalesMan"], values="Sale_amt")
print(table.query('Manager == ["Timothy"]'))
```
